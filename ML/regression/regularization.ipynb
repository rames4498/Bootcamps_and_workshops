```python
col = ['Suburb', 'Rooms', 'Type', 'Price', 'Method', 'SellerG', 'Distance', 'Bedroom2', 'Bathroom', 'Car','Landsize', 'BuildingArea', 'CouncilArea', 'Regionname', 'Propertycount']
```

```python
dataf  =  df[col]
dataf.head()
```

```python
dataf.shape
```

```python
dataf.isna().sum()
```

```python
null_fill  =  ['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
dataf[null_fill]   =  dataf[null_fill].fillna(0)
```

```python
dataf.isna().sum()
```

```python
dataf['Landsize'] = dataf['Landsize'].fillna(dataf.Landsize.mean())
dataf['BuildingArea'] = dataf['BuildingArea'].fillna(dataf.Landsize.mean())
```

```python
dataf.dropna(inplace=True)
dataf
```

```python
dataf.isna().sum()
```

```python
# encoding 

dataf =  pd.get_dummies(dataf, drop_first=True)
dataf
```

```python
# train and test split
```

```python
X  =  dataf.drop('Price', axis =1)
y  =  dataf['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y =  train_test_split(X,y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model =  LinearRegression().fit(train_X, train_y)
```

```python
model.score(test_X,test_y)
```

```python
model.score(train_X,train_y)
```

```python
# regularization
```

```python
from sklearn import linear_model

lass_model  =  linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)

lass_model.fit(train_X, train_y)
```

```python
lass_model.score(test_X, test_y)
```

```python
lass_model.score(train_X, train_y)
```

```python
from sklearn.linear_model import Ridge

Ridge_model  =  linear_model.Ridge(alpha=50, max_iter=100, tol=0.1)

Ridge_model.fit(train_X, train_y)
```

```python
Ridge_model.score(test_X, test_y)
```
