```python
import pandas as pd
import numpy as np
import re
```

```python
df= pd.read_csv('mobile_price_data.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
# this the function of converting there dtype
def func(df,column_name,string_to_be_replaced,convert_type,sndf=None):
    if sndf !=None:
        df[column_name]=df[column_name].str.replace(sndf,'')
    df[column_name]=df[column_name].str.replace(string_to_be_replaced,'').astype(convert_type)
    x = column_name+'('+string_to_be_replaced+')'
    df.rename({column_name: x},axis = 1,inplace = True) # rename our column name with there unit
```

```python
# Extractiing data from disp_size column and putting it into anpther column
df['disp_size(cm)'] = df['disp_size'].str.extract('([0-9.]+)')
df['disp_size(cm)'] = df['disp_size(cm)'].astype(float)
df.drop('disp_size',axis=1,inplace=True)
```

```python
x = '\u20B9'  # Rupees sign
```

```python
func(df,'mp_speed',' GHz',float)   # for MP_speed
func(df,'int_memory',' GB',int)   # for memory
func(df,'mobile_price',x,int,',')    # converting our mobile price in interger
func(df,'ram',' GB',int)
func(df,'battery_power','mAh',int)
func(df,'mob_width','mm',float)
func(df,'mob_height','mm',float)
func(df,'mob_depth','mm',float)
func(df,'mob_weight',' g',float)
```
