```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("Classified Data",index_col=0)
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler  =  StandardScaler()
```

```python
df.drop("TARGET CLASS",axis=1)
```

```python
df
```

```python
scaler.fit(df.drop("TARGET CLASS",axis=1))
```

```python
features_scaled = scaler.transform(df.drop("TARGET CLASS",axis=1))
```

```python
newDf  = pd.DataFrame(features_scaled, columns=df.columns[:-1])
newDf
```

```python
import seaborn as sns
sns.pairplot(df, hue  = 'TARGET CLASS')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test =  train_test_split(features_scaled, df['TARGET CLASS'],test_size = 0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn =  KNeighborsClassifier(n_neighbors=23)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions =  knn.predict(X_test)

# predictions should be closer to Y_test values
```

```python
# performance metrics 

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
# we have to choose our nearest neighbors or K value
```

```python
accuracy = []

for i in range(1,45):
    
    knn =  KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(knn,newDf, df['TARGET CLASS'],cv =10)
    accuracy.append(score.mean())
```

```python
error_rate = []

for i in range(1,45):
    
    knn =  KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    predict  = knn.predict(X_test)
    error_rate.append(np.mean(predict != y_test))
```

```python
plt.figure(figsize=(8,6))

plt.plot(range(1,45),accuracy, color= 'red', linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)

plt.title('Error Rate vs K values')

plt.xlabel('K')

plt.ylabel('Error Rate')


```
