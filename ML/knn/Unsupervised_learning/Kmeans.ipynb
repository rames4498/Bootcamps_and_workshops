```python
import numpy as np
import pandas as pd
# income dataset
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df  =  pd.read_csv('income.csv')
df.head(10)
```

```python
df.info()
```

```python
plt.scatter(df.Age, df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
model =  KMeans(n_clusters=3)
y_predict = model.fit_predict(df[['Age','Income($)']])
y_predict
```

```python
df['cluster'] = y_predict

df.head(22)
```

```python
model.cluster_centers_
```

```python
#df[df.cluster == 1]

df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]

plt.scatter(df0.Age, df0['Income($)'],color='red')
plt.scatter(df1.Age, df1['Income($)'],color='green')
plt.scatter(df2.Age, df2['Income($)'],color='blue')

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='purple')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
# minmax scaler

scaler  =  MinMaxScaler()

scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df.head(22)
```

```python
plt.scatter(df.Age, df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
model =  KMeans(n_clusters=3)
y_predict = model.fit_predict(df[['Age','Income($)']])
y_predict
```

```python
df['cluster'] = y_predict

df.head(22)
```

```python
#df[df.cluster == 1]

df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]

plt.scatter(df0.Age, df0['Income($)'],color='red')
plt.scatter(df1.Age, df1['Income($)'],color='green')
plt.scatter(df2.Age, df2['Income($)'],color='blue')

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='purple')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
#elbow method

sse  =  []

rang  =  range(1,22)


for k in  rang:
    model  = KMeans(n_clusters = k)
    model.fit(df[['Age','Income($)']])
    sse.append(model.inertia_)



```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(rang, sse)
```
